```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("Stock_data.csv")

df
```

```python
#1) Ask user for start date and end date between that period.
# calculate 12-period EMA and 26-period EMA are two moving averages.
#Calculate them and plot These two with Close price in a line chart

# Convert the 'Gmt time' column to datetime format
df['Gmt time'] = pd.to_datetime(df['Gmt time'])

# Ask the user for start and end dates
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

# Filter the dataset based on user input
mask = (df['Gmt time'] >= start_date) & (df['Gmt time'] <= end_date)
filtered_df = df.loc[mask]

# Calculate 12-period EMA and 26-period EMA
filtered_df['12EMA'] = filtered_df['Close'].ewm(span=12, adjust=False).mean()
filtered_df['26EMA'] = filtered_df['Close'].ewm(span=26, adjust=False).mean()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['Gmt time'], filtered_df['Close'], label='Close Price', linewidth=2)
plt.plot(filtered_df['Gmt time'], filtered_df['12EMA'], label='12-period EMA', linestyle='--', linewidth=2)
plt.plot(filtered_df['Gmt time'], filtered_df['26EMA'], label='26-period EMA', linestyle='--', linewidth=2)

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price with 12 and 26-period EMAs')
plt.legend()

# Show the plot
plt.show()
```

```python
#2) Ask user for start date and end date between that period calculate. 12-period SMA and 26-period SMA are two moving averages. Calculate them and plot These two with Close price in a line chart

start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

# Filter the data based on user input
mask = (df['Gmt time'] >= start_date) & (df['Gmt time'] <= end_date)
filtered_data = df.loc[mask]

# Calculate 12-period SMA and 26-period SMA
sma_12 = filtered_data['Close'].rolling(window=12).mean()
sma_26 = filtered_data['Close'].rolling(window=26).mean()

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(filtered_data['Gmt time'], filtered_data['Close'], label='Close Price', color='blue')
plt.plot(filtered_data['Gmt time'], sma_12, label='12-period SMA', color='green')
plt.plot(filtered_data['Gmt time'], sma_26, label='26-period SMA', color='red')

# Customize the plot
plt.title('Stock Data with 12-period and 26-period SMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()


```

```python
#a)  compare SMA 12 EMA 12 and CLOSE
#b) compare SMA 26 EMA 26 and CLOSE


start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

# Filter the data based on user input
mask = (df['Gmt time'] >= start_date) & (df['Gmt time'] <= end_date)
filtered_data = df.loc[mask]

# Calculate 12-period SMA, EMA, and 26-period SMA, EMA
sma_12 = filtered_data['Close'].rolling(window=12).mean()
ema_12 = filtered_data['Close'].ewm(span=12, adjust=False).mean()

sma_26 = filtered_data['Close'].rolling(window=26).mean()
ema_26 = filtered_data['Close'].ewm(span=26, adjust=False).mean()

# Plot the data for SMA 12, EMA 12, and CLOSE
plt.figure(figsize=(10, 6))
plt.plot(filtered_data['Gmt time'], filtered_data['Close'], label='Close Price', color='blue')
plt.plot(filtered_data['Gmt time'], sma_12, label='12-period SMA', color='green')
plt.plot(filtered_data['Gmt time'], ema_12, label='12-period EMA', color='red')

# Customize the plot
plt.title('Comparison of 12-period SMA, 12-period EMA, and Close Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()

# Plot the data for SMA 26, EMA 26, and CLOSE
plt.figure(figsize=(10, 6))
plt.plot(filtered_data['Gmt time'], filtered_data['Close'], label='Close Price', color='blue')
plt.plot(filtered_data['Gmt time'], sma_26, label='26-period SMA', color='green')
plt.plot(filtered_data['Gmt time'], ema_26, label='26-period EMA', color='red')

# Customize the plot
plt.title('Comparison of 26-period SMA, 26-period EMA, and Close Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Ask the user for start and end dates
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

# Filter the dataset based on user input
mask = (df['Gmt time'] >= start_date) & (df['Gmt time'] <= end_date)
filtered_df = df.loc[mask]

# Calculate 12-period EMA and 26-period EMA
filtered_df['12EMA'] = filtered_df['Close'].ewm(span=12, adjust=False).mean()
filtered_df['26EMA'] = filtered_df['Close'].ewm(span=26, adjust=False).mean()

# Calculate MACD (12EMA - 26EMA)
filtered_df['MACD'] = filtered_df['12EMA'] - filtered_df['26EMA']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['Gmt time'], filtered_df['12EMA'], label='12-period EMA', linestyle='--', linewidth=2)
plt.plot(filtered_df['Gmt time'], filtered_df['26EMA'], label='26-period EMA', linestyle='--', linewidth=2)
plt.plot(filtered_df['Gmt time'], filtered_df['MACD'], label='MACD', linewidth=2)

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Price/MACD')
plt.title('Stock Price with 12 and 26-period EMAs and MACD')
plt.legend()

# Show the plot
plt.show()
```
