```python
import random

def print_board(board):
  print("The board is: \n")
  print(" {} | {} | {} ".format(board[0][0],board[0][1],board[0][2]))
  print("-----------")
  print(" {} | {} | {} ".format(board[1][0],board[1][1],board[1][2]))
  print("-----------")
  print(" {} | {} | {} ".format(board[2][0],board[2][1],board[2][2]))

def check_win(board, player):
    for row in board:
        if all(cell == player for cell in row):
            return True
    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True
    return False

def is_draw(board):
    return all(cell != " " for row in board for cell in row)

def get_empty_cells(board):
    return [(row, col) for row in range(3) for col in range(3) if board[row][col] == " "]

def computer_move(board, symbol):
    empty_cells = get_empty_cells(board)
    return random.choice(empty_cells)

def main():
    print("Welcome to Tic-Tac-Toe!")
    player_name = input("Enter your name: ")
    player_symbol = input("Choose X or O: ").upper()
    computer_symbol = "X" if player_symbol == "O" else "O"

    players = [player_name, "Computer"]
    symbols = [player_symbol, computer_symbol]

    turn = random.choice([0, 1])
    print(f"{players[turn]} will go first!")

    board = [[" " for _ in range(3)] for _ in range(3)]

    while True:
        print_board(board)

        row, col = -1, -1
        if turn == 0:
            while row < 0 or board[row][col] != " ":
                row = int(input(f"{players[turn]}, enter row (0, 1, 2): "))
                col = int(input(f"{players[turn]}, enter column (0, 1, 2): "))
        else:
            row, col = computer_move(board, symbols[turn])
            print(f"{players[turn]} chose row {row}, column {col}.")

        if board[row][col] == " ":
            board[row][col] = symbols[turn]
            if check_win(board, symbols[turn]):
                print_board(board)
                print(f"Congratulations! {players[turn]} wins!")
                break
            elif is_draw(board):
                print_board(board)
                print("It's a draw!")
                break
            turn = 1 - turn  # Switch turns
    play_again = input("Do you want to play again? (yes/no): ")
    if play_again.lower() == "yes":
        main()
    else:
        print("Thank you for playing!")

if __name__ == "__main__":
    main()
```
