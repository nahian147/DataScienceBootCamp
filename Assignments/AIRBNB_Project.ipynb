```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
#1) Calculate How many properties in each neighbourhood group

properties_by_neighbourhood_group = df.groupby('neighbourhood_group')['id'].count()


print(properties_by_neighbourhood_group)
```

```python
#2) Which Neighbour Hood Group Has highest prices use bar chart to show this for all neighbourhood groups


prices_by_neighbourhood = df.groupby('neighbourhood_group')['price'].mean()
highest_price_neighbourhood_group = prices_by_neighbourhood.idxmax()
#Neighborhood group with the highest average price
print("The neighbourhood group with the highest prices is:", highest_price_neighbourhood_group)


# Bar chart
plt.figure(figsize=(10, 6))
prices_by_neighbourhood.plot(kind='bar', color='blue')
plt.title('Average Prices by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()




```

```python
#3) Which Neighbourhood group has lowest prices use bar chart to show this for all neighbourhood groups


prices_by_neighbourhood = df.groupby('neighbourhood_group')['price'].sum()
lowest_price_neighbourhood_group = prices_by_neighbourhood.idxmin()

#Neighborhood group with the lowest average price
print("The neighbourhood group with the lowest prices is:" , lowest_price_neighbourhood_group)




# Bar chart
plt.figure(figsize=(10, 6))
prices_by_neighbourhood.plot(kind='bar', color='blue')
plt.title('Average Prices by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#4) Highest and lowest price airbnb in each Neighbour Hood Group



# Find the indices of the rows with the highest and lowest prices in each Neighbourhood Group
highest_price_indices = df.groupby('neighbourhood_group')['price'].idxmax()
lowest_price_indices = df.groupby('neighbourhood_group')['price'].idxmin()

# Extract the corresponding rows
highest_price_listings = df.loc[highest_price_indices]
lowest_price_listings = df.loc[lowest_price_indices]

# Display the result
print("Highest Price Listings in Each Neighbourhood Group:")
print(highest_price_listings[['neighbourhood_group', 'neighbourhood', 'price']])

print("\nLowest Price Listings in Each Neighbourhood Group:")
print(lowest_price_listings[['neighbourhood_group', 'neighbourhood', 'price']])
```

```python

#5) Which Host has highest count of properties

host_property_count = df.groupby('host_id')['id'].count()

# Find the host with the highest count of properties
highest_property_host_id = host_property_count.idxmax()
highest_property_count = host_property_count.max()

# Display the result
print(f"The host with the highest count of properties is Host ID {highest_property_host_id} with {highest_property_count} properties.")
```

```python
#6) Which neighbourhood is most expensive and cheapest top 10 barchart
average_prices_by_neighbourhood = df.groupby('neighbourhood')['price'].mean()

# Find the top 10 most expensive neighborhoods
top_10_expensive_neighbourhoods = average_prices_by_neighbourhood.nlargest(10)

# Find the top 10 cheapest neighborhoods
top_10_cheapest_neighbourhoods = average_prices_by_neighbourhood.nsmallest(10)

# Create a bar chart for the top 10 most expensive neighborhoods
plt.figure(figsize=(12, 6))
top_10_expensive_neighbourhoods.plot(kind='bar', color='salmon')
plt.title('Top 10 Most Expensive Neighbourhoods')
plt.xlabel('Neighbourhood')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()

# Create a bar chart for the top 10 cheapest neighborhoods
plt.figure(figsize=(12, 6))
top_10_cheapest_neighbourhoods.plot(kind='bar', color='lightblue')
plt.title('Top 10 Cheapest Neighbourhoods')
plt.xlabel('Neighbourhood')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
#7) Count different type of room_type and do a pie chart showing % of each room type in overall dataset

room_type_counts = df['room_type'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightblue', 'lightgreen', 'lightyellow'])
plt.title('Distribution of Room Types in Airbnb Listings')
plt.show()
```

```python
#8) scatter plot between longitutude and latitude for different neighbourhood groups

colors = {'Brooklyn': 'blue', 'Manhattan': 'red', 'Queens': 'green', 'Bronx': 'purple', 'Staten Island': 'orange'}

# Create a scatter plot
plt.figure(figsize=(12, 8))
for group, data in df.groupby('neighbourhood_group'):
    plt.scatter(data['longitude'], data['latitude'], label=group, color=colors[group], alpha=0.5)

# Set labels and title
plt.title('Scatter Plot of Longitude vs Latitude for Different Neighbourhood Groups')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
#9) scatter plot between longitutude and latitude for different room_types
colors = {'Entire home/apt': 'blue', 'Private room': 'red', 'Shared room': 'green', 'Hotel room': 'purple'}

# Create a scatter plot
plt.figure(figsize=(12, 8))
for room_type, data in df.groupby('room_type'):
    plt.scatter(data['longitude'], data['latitude'], label=room_type, color=colors[room_type], alpha=0.5)

# Set labels and title
plt.title('Scatter Plot of Longitude vs Latitude for Different Room Types')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
#10) room_type vs price bar chart for the mean and highest and lowest


# Calculate the mean, highest, and lowest prices for each room type
room_type_stats = df.groupby('room_type')['price'].agg(['mean', 'max', 'min'])

# Plotting the bar chart
plt.figure(figsize=(12, 8))
room_type_stats.plot(kind='bar', color=['lightblue', 'lightcoral', 'lightgreen'], edgecolor='black')
plt.title('Room Type vs Price Statistics')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.xticks(rotation=0)
plt.legend(['Mean Price', 'Highest Price', 'Lowest Price'])
plt.show()
```

```python
#11)Do a similar plot to this

df_subset = df[['room_type','neighbourhood_group']]

# Create a cross-tabulation of room types and neighborhood groups
cross_tab = pd.crosstab(df_subset['neighbourhood_group'],df_subset['room_type'])

# Transpose the cross-tabulation
cross_tab = cross_tab.T

# Plotting the grouped bar chart
plt.figure(figsize=(12, 8))
cross_tab.plot(kind='bar', color=['lightblue', 'lightcoral', 'lightgreen', 'lightyellow'], edgecolor='black')
plt.title('Room Type Distribution in Neighborhood Groups')
plt.xlabel('Room types')
plt.ylabel('Number of Listings')
plt.xticks(rotation=0)
plt.legend(title='Neighborhood group')
plt.show()
```

```python
df_new = df[['neighbourhood_group', 'latitude', 'longitude', 'room_type',
         'minimum_nights', 'number_of_reviews', 'reviews_per_month',
         'calculated_host_listings_count','price']]

df_new.head()
```

Splitting Independent X and Dependant Y

```python
X = df_new.iloc[:, :-1].values
y = df_new.iloc[:, -1].values
X,y
```

```python
X[0]
```

#**Applying one hot encoder in 0th and 3rd column neighborhood_group and roomtype**


```python
#Applying one hot encoder in 0th and 3rd column neighborhood_group and roomtype
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0,3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
X
```

```python
X[0]
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#**LINEAR REGRESSION**

## Training the Multiple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score


mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred.round())

print('Mean Squared Error', mse)
print('Mean Absolute Error', mae)
print('Accuracy:', accuracy)
```
