```python
#2) Create a Dictionary with the 10 countries that have the most 'Quantity' ordered except UK

df_notUK = df_cleaned[df_cleaned['Country'] != 'United Kingdom']
country_quantities = {}


for index, row in df_cleaned.iterrows():
    country = row['Country']
    quantity = row['Quantity']

    if country in country_quantities:
        country_quantities[country] += quantity
    else:
        country_quantities[country] = quantity

sorted_country_quantities = dict(sorted(country_quantities.items(), key=lambda x: x[1], reverse=True)[:10])

print(sorted_country_quantities)
```

```python
#2 Alternative: by group by

top_countries = df_cleaned[df_cleaned['Country'] != 'UK'].groupby('Country')['Quantity'].sum().nlargest(10).to_dict()

print(top_countries )
```

```python
#3)  Find Total Sale between two dates entered by user
df_copy = df_cleaned.copy()
df_copy['InvoiceDate'] = pd.to_datetime(df_copy['InvoiceDate'])

start= input("Enter the start date (YYYY-MM-DD): ")
end= input("Enter the end date (YYYY-MM-DD): ")

start_date = pd.to_datetime(start)
end_date = pd.to_datetime(end)

filtered_df = df_copy[(df_copy['InvoiceDate'] >= start_date) & (df_copy['InvoiceDate'] <= end_date)]

total = (filtered_df['Quantity'] * filtered_df['UnitPrice']).sum()

print("Total Sale between", start, "and", end, "is:", total)
```

```python
#4) Find each Countries Total SALE to find which Countries is highest potential customer
country_sales = df.groupby('Country')['Quantity'].sum()
highest_potential_customer = country_sales.idxmax()

print(country_sales)
print(highest_potential_customer)
```

```python
#5) Which Customer Id has highest purchase in quantity and total and find out the country of it

highest_customer_id = None
highest_purchase_quantity = 0
highest_purchase_total = 0
highest_purchase_country = None


for index, row in df_cleaned.iterrows():
    customer_id = row['CustomerID']
    quantity = row['Quantity']
    total_purchase = quantity * row['UnitPrice']
    country = row['Country']

    if total_purchase > highest_purchase_total:
        highest_purchase_quantity = quantity
        highest_purchase_total = total_purchase
        highest_customer_id = customer_id
        highest_purchase_country = country

print("Customer with the Highest Purchase:")
print("Customer ID:", highest_customer_id)
print("Total Quantity:", highest_purchase_quantity)
print("Total Purchase:", highest_purchase_total)
print("Country:", highest_purchase_country)
```

```python
# Alternative: Which Customer ID has the highest purchase in quantity and total, and find out the country of it
customer_purchase = df_cleaned.groupby('CustomerID').agg({'Quantity': 'sum', 'UnitPrice': 'sum'})
max_quantity_customer = customer_purchase['Quantity'].idxmax()
max_quantity_customer_country = df_cleaned[df_cleaned['CustomerID'] == max_quantity_customer]['Country'].iloc[0]

print(max_quantity_customer)

print(max_quantity_customer_country )
```

```python
#6) Which Customer Id has lowest purchase in quantity and total and find out the country of it

lowest_customer_id = None
lowest_purchase_quantity = float('inf')
lowest_purchase_total = float('inf')
lowest_purchase_country = None


for index, row in df_cleaned.iterrows():
    customer_id = row['CustomerID']
    quantity = row['Quantity']

    total_purchase = quantity * row['UnitPrice']
    country = row['Country']

    if total_purchase < lowest_purchase_total:
        lowest_purchase_quantity = quantity
        lowest_purchase_total = total_purchase
        lowest_customer_id = customer_id
        lowest_purchase_country = country

print("Customer with the lowest Purchase:")
print("Customer ID:", lowest_customer_id)
print("Total Quantity:", lowest_purchase_quantity)
print("Total Purchase:", lowest_purchase_total)
print("Country:", lowest_purchase_country)
```

```python
#7) Find the biggest order in Sale (Cost), which country was the order from, and find the total purchase (Cost) done by that Customer ID
df_cleaned['TotalCost'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']
biggest_order = df_cleaned[df_cleaned['TotalCost'] == df_cleaned['TotalCost'].max()]
biggest_order_country = biggest_order['Country'].iloc[0]
customer_total_purchase = customer_purchase.loc[biggest_order['CustomerID'].iloc[0], 'UnitPrice']
```

```python
#8: Find how many different customers are available for each country, save the count in a dictionary, and return countries with the highest and lowest customer count
country_customer_counts = df_cleaned.groupby('Country')['CustomerID'].nunique().to_dict()
max_customer_count_country = max(country_customer_counts, key=country_customer_counts.get)
min_customer_count_country = min(country_customer_counts, key=country_customer_counts.get)
```

```python
#9) Find which stock code was ordered the most by each country
country_stock_most_ordered = df_cleaned.groupby(['Country', 'StockCode'])['Quantity'].sum().reset_index()


most_ordered_stock_by_country = country_stock_most_ordered.groupby('Country').apply(lambda x: x.nlargest(1, 'Quantity')).reset_index(drop=True)


print(most_ordered_stock_by_country)
```
