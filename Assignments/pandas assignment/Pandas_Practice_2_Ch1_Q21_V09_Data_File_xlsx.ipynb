```python
import pandas as pd
df=pd.read_excel("Ch1_Q21_V09_Data_File.xlsx")
```

```python
df=df.iloc[:,:5]
```

```python
#a.
print("How many observations have x4 values are less than 3?")

print((df['x4'] < 3).sum())
```

```python
#b. b. Sort the data by x1 in ascending order. What are the x1, x2, x3, and x4 values of the first observation after the data are sorted?  (For x1, round your answer to four decimal places.)

sorted_df = df.sort_values(by='x1', ascending=True)


print("x1:", round(sorted_df.iloc[0]['x1'],4))
print("x2:", sorted_df.iloc[0]['x2'])
print("x3:", sorted_df.iloc[0]['x3'])
print("x3:", sorted_df.iloc[0]['x4'])
```

```python
#c. Sort the data by x1 in descending order. What are the x1, x2, x3, and x4 values of the first observation after the data are sorted? (For x1, round your answer to four decimal places.)


sorted_df = df.sort_values(by='x1', ascending=False)


print("x1:", round(sorted_df.iloc[0]['x1'],4))
print("x2:", sorted_df.iloc[0]['x2'])
print("x3:", sorted_df.iloc[0]['x3'])
print("x4:", sorted_df.iloc[0]['x4'])
```

```python
#d. How many missing values are there in x1, x2, x3, and x4?
missing_x1 = df['x1'].isna().sum()
missing_x2 = df['x2'].isna().sum()
missing_x3 = df['x3'].isna().sum()
missing_x4 = df['x4'].isna().sum()

print("Missing in x1:", missing_x1)
print("Missing in x2:", missing_x2)
print("Missing in x3:", missing_x3)
print("Missing in x4:", missing_x4)
```

```python
#e. How many observations are there in each category in x4?
category_counts = df['x4'].value_counts()
print(category_counts)
```
