```python
#a. Subset the data set to include only x2, x3, and x4. How many missing values are there in these three variables?
subset_df = df[['x2', 'x3', 'x4']]

missing_values = subset_df.isna().sum()

print("missing_values:")
print(missing_values)
```

```python
sheet_name = 1  # Replace with the name of the sheet you want to read
df_s2 = pd.read_excel('Ch1_Q23_V02_Data_File.xlsx', sheet_name)
df_s2
```

```python
df_s2=df_s2.iloc[:,:5]
df_s2
```

```python
#a. Count the number of males and females in the data.

num_males = df_s2[df_s2['Sex'] == 'M']['Individual'].count()
num_females = df_s2[df_s2['Sex'] == 'F']['Individual'].count()

print("Males:")
print(num_males)

print("Females:")
print(num_females)
```

```python
# b. Calculate the percentage of males and females who are married


total_males = num_males
total_females = num_females

num_married_males = df_s2[(df_s2['Sex'] == 'M') & (df_s2['Married'] == 'Y')]['Individual'].count()
num_married_females = df_s2[(df_s2['Sex'] == 'F') & (df_s2['Married'] == 'Y')]['Individual'].count()

percentage_married_males = (num_married_males / total_males) * 100
percentage_married_females = (num_married_females / total_females) * 100
```

```python


print("Percentage of married males:", round(percentage_married_males, 2), "%")
print("Percentage of married females:", round(percentage_married_females, 2), "%")
```

```python
# c. Find the number of married males among the 10 individuals with the highest income

top_10_highest_income = df_s2.nlargest(10, 'Income')
num_married_males_top_10 = top_10_highest_income[(top_10_highest_income['Sex'] == 'M') & (top_10_highest_income['Married'] == 'Y')]['Individual'].count()

print("c. Number of married males among the 10 individuals with the highest income:", num_married_males_top_10)
```
