```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("Ch3_Q25_Data_File.xlsx")

df
```

```python
df=df.iloc[:,:2]
df
```

```python
# a. Calculate the 25th, 50th, and 75th percentiles
percentiles = df['PE_Ratio'].quantile([0.25, 0.5, 0.75])

Q1 = round(percentiles[0.25], 2)
median = round(percentiles[0.5], 2)
Q3 = round(percentiles[0.75], 2)

print("25th Percentile (Q1):", Q1)
print("50th Percentile (Median):", median)
print("75th Percentile (Q3):", Q3)

# b-1. Create a box plot
plt.boxplot(df['PE_Ratio'])
plt.title("Box Plot of P/E Ratios")
plt.ylabel("P/E Ratio")
plt.show()

# b-2. Determine if there are any outliers
q1 = df['PE_Ratio'].quantile(0.25)
q3 = df['PE_Ratio'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['PE_Ratio'] < lower_bound) | (df['PE_Ratio'] > upper_bound)]

if not outliers.empty:
    print("b-2. There are outliers.")
    print(outliers)
else:
    print("b-2. There are no outliers.")

# b-3. Determine if the distribution is symmetric
plt.hist(df['PE_Ratio'], bins=10)
plt.title("Histogram of P/E Ratios")
plt.xlabel("P/E Ratio")
plt.ylabel("Frequency")
plt.show()
```
