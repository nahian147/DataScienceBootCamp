```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("Ch3_Q45_Data_File.xlsx")

df
```

```python
df=df.iloc[:,:2]
df
```

```python
mean_rent = df['Rent'].mean()
std_deviation_rent = df['Rent'].std()


mean_footage = df['Footage'].mean()
std_deviation_footage = df['Footage'].std()

coefficient_of_variation_rent = std_deviation_rent / mean_rent
coefficient_of_variation_footage = std_deviation_footage / mean_footage

print("a. Mean Monthly Rent:", round(mean_rent, 2))
print("   Standard Deviation for Monthly Rent:", round(std_deviation_rent, 2))
print("b. Mean Footage:", round(mean_footage, 2))
print("   Standard Deviation for Footage:", round(std_deviation_footage, 2))

if coefficient_of_variation_rent > coefficient_of_variation_footage:
    print("c. The monthly rent data exhibit greater relative dispersion.")
else:
    print("c. The footage data exhibit greater relative dispersion.")
```
