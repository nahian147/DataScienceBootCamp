```python
import pandas as pd
```

```python
df=pd.read_excel("Ch1_Q20_V12_Data_File.xlsx")
```

```python
df=df.iloc[:,:4]
```

```python
df
```

```python
#a.
print("How many observations have x1 values greater than 30?")

print((df['x1'] > 30).sum())

#or
#print(len(df[df['x1']>30]))
```

```python
#b. Sort the data by x1 in ascending order. What are the x1, x2, and x3 values of the first observation after the data are sorted?

sorted_df = df.sort_values(by='x1', ascending=True)


print("x1:", sorted_df.iloc[0]['x1'])
print("x2:", sorted_df.iloc[0]['x2'])
print("x3:", sorted_df.iloc[0]['x3'])
```

```python
sorted_df = df.sort_values(by='x1', ascending=False)


print("x1:", sorted_df.iloc[0]['x1'])
print("x2:", sorted_df.iloc[0]['x2'])
print("x3:", sorted_df.iloc[0]['x3'])
```

```python
#d. How many missing values are there in x1, x2, and x3?
print('Missing val in x1:', df['x1'].isna().sum())
print('Missing val in x2:', df['x2'].isna().sum())
print('Missing val in x3:', df['x3'].isna().sum())
```
