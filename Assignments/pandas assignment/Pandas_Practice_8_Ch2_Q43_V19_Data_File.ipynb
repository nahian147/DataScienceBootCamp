```python
import pandas as pd

import matplotlib.pyplot as plt

df=pd.read_excel("Ch2_Q43_V19_Data_File.xlsx")

df
```

```python
df=df.iloc[:,:3]
df
```

```python
plt.figure(figsize=(12, 6))  # Adjust the figure size if needed
plt.plot(df['Date'], df['AMZN'], label='Amazon (AMZN)', color='blue')
plt.plot(df['Date'], df['GOOG'], label='Alphabet (GOOG)', color='green')

plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Amazon vs. Alphabet Stock Prices Over Time')

plt.legend()

plt.grid(True)
plt.show()
```
