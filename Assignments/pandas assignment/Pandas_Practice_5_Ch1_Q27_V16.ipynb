```python
import pandas as pd

df=pd.read_excel("Ch1_Q27_V16_Data_File.xlsx")

df
```

```python
df=df.iloc[:,:5]
df
```

```python
num_males = df[df['Sex'] == 'M']['Individual'].count()
num_females = df[df['Sex'] == 'F']['Individual'].count()


total_males = num_males
total_females = num_females

num_married_males = df[(df['Sex'] == 'M') & (df['Married'] == 'Y')]['Individual'].count()
num_married_females = df[(df['Sex'] == 'F') & (df['Married'] == 'Y')]['Individual'].count()

percentage_married_males = (num_married_males / total_males) * 100
percentage_married_females = (num_married_females / total_females) * 100


top_10_highest_income = df.nlargest(10, 'Income')
num_married_males_top_10 = top_10_highest_income[(top_10_highest_income['Sex'] == 'M') & (top_10_highest_income['Married'] == 'Y')]['Individual'].count()


print("a. Number of males:", num_males)
print("   Number of females:", num_females)
print("b. Percentage of married males:", round(percentage_married_males, 2), "%")
print("   Percentage of married females:", round(percentage_married_females, 2), "%")
print("c. Number of married males among the 10 individuals with the highest income:", num_married_males_top_10)
```
