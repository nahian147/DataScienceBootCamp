```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("Ch3_Q43_Data_File.xlsx")

df
```

```python
df=df.iloc[:,:2]
df
```

```python
minimum_expenditure = df['Expenditures'].min()
maximum_expenditure = df['Expenditures'].max()

mean_expenditure = df['Expenditures'].mean()
median_expenditure = df['Expenditures'].median()


variance_expenditure = df['Expenditures'].var()
std_deviation_expenditure = df['Expenditures'].std()

print("a. Minimum Expenditure:", minimum_expenditure)
print("   Maximum Expenditure:", maximum_expenditure)
print("b. Mean Expenditure:", round(mean_expenditure, 2))
print("   Median Expenditure:", round(median_expenditure, 2))
print("c. Variance:", round(variance_expenditure, 2))
print("   Standard Deviation:", round(std_deviation_expenditure, 2))
```
