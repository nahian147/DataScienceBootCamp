```python
import pandas as pd
#import matplotlib.pyplot as plt
df=pd.read_excel("Ch03_Q71_V03_Data-File.xlsx")

df
```

```python
df=df.iloc[:5,:2]
df
```

```python
# a. Calculate the covariance between x and y
covariance = df['x'].cov(df['y'])
print("a. Covariance between x and y:", round(covariance, 2))

# b. Calculate the correlation coefficient between x and y
correlation_coefficient = df['x'].corr(df['y'])
print("b. Correlation coefficient between x and y:", round(correlation_coefficient, 2))
```
