```python
import pandas as pd

df=pd.read_excel("Ch2_Q15_V09_Data_File.xlsx")

df
```

```python
df=df.iloc[:,:3]
df
```

```python
contingency_table = pd.crosstab(df['Sex'], df['Feasible'])
print("Contingency Table:")
print(contingency_table)

# a-2. Calculate the number of female students
num_female_students = df[df['Sex'] == 'female']['Student'].count()
print("\na-2. Number of female students:", num_female_students)

# a-3. Calculate the number of students who feel it's feasible
num_feasible_students = df[df['Feasible'] == 'yes']['Student'].count()
print("\na-3. Number of students who feel it's feasible:", num_feasible_students)

# b-1. Calculate the likelihood for male students
num_male_students = df[df['Sex'] == 'male']['Student'].count()
likelihood_male = (contingency_table.loc['male', 'yes'] / num_male_students) * 100
print(f"\nb-1. Likelihood of a male student feeling it's feasible: {likelihood_male:.2f}%")

# b-2. Calculate the likelihood for female students
likelihood_female = (contingency_table.loc['female', 'yes'] / num_female_students) * 100
print(f"\nb-2. Likelihood of a female student feeling it's feasible: {likelihood_female:.2f}%")
```
