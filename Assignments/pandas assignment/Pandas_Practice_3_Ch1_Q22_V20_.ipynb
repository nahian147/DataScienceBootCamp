```python
import pandas as pd

df=pd.read_excel("Ch1_Q22_V20_Data_File.xlsx")

df
```

```python
df=df.iloc[:,:4]
df
```

```python


#a. Subset the data set to include only observations that have a date on or after May 1, 1975 for x3. How many observations are in the subset data?

df['x3'] = pd.to_datetime(df['x3'])

subset_data = df[df['x3'] >= '1975-05-01']

print("Observations that have a date on or after May 1, 1975 for x3")
print(subset_data)


print("How many observations are in the subset data?")
print(len(subset_data))

```

```python
#b.Subset the original data set based on the binary 1/0 values for x4.

subset_1 = df[df['x4']==1]
subset_0 = df[df['x4']==0]

print("Subset For 1")
print(subset_1)

print("Subset For O")
print(subset_0)

#How many observations are in each of the two subsets?
print("observations For Subset1")
print(len(subset_1))

print("observations For Subset1")
print(len(subset_0))
```
