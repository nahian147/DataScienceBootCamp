```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("Ch2_Q57_Data_File.xlsx")

df
```

```python
df=df.iloc[:,:4]
df
```

```python
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
plt.scatter(df['Education'], df['Wage'], marker='o', color='blue', alpha=0.5)
plt.xlabel('Years of Education')
plt.ylabel('Hourly Wage ($)')
plt.title(' Wage vs. Education')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Adjust the figure size if needed
plt.scatter(df['Age'], df['Wage'], marker='o', color='green', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Hourly Wage ($)')
plt.title('WAGE vs. AGE')
plt.grid(True)

# Display the scatterplot
plt.show()
```
